Ejercicio 1
Define el conjunto C con los enteros de -10 a 10 y luego obtener el subconjunto D
con los números enteros negativos mayores que -6.

```python
C = {x for x in range(-10, 11)}  # Conjunto de enteros de -10 a 10
D = {x for x in C if x < 0 and x > -6}  # Obtenemos los negativos mayores que -6
print("Conjunto C:", C)
print("Conjunto D (negativos mayores que -6):", D)
```

Ejercicio 2. Números impares menores que 15
Define el conjunto E con los números naturales del 1 al 20 y luego obtener el
subconjunto F con los números impares menores que 15.

```python
E = {x for x in range(1, 21)} # Conjunto de los numeros del 1 al 20
F = {x for x in E if x%2!=0 and x<15} #Evaluamos si es impar y menor a 15
print("Conjunto E:", E)
print("Conjunto F (números impares menores que 15):",F)
```

Ejercicio 3. Fracciones unitarias menores a 0.2
Definir el conjunto G con los valores racionales de la forma 1/n donde n toma
valores de 1 a 10, y luego obtener el subconjunto H con los valores menores a 0.2.

```python
G = {1/n for n in range(1, 11)}  # Conjunto de valores 1/n, por eso se escribe asi al inicio, en el rango de 1 a 10
H = {x for x in G if x < 0.2}  # Obtenemos las fraciones del conjunto anterior  menores a 0.2
print("Conjunto G:", G)
print("Conjunto H (fracciones < 0.2):", H)
```

Ejercicio 4. Números primos menores que 30
Definir el conjunto I con los números naturales del 1 al 30, y luego obtener el
subconjunto J con los números primos.

```python
I = {x for x in range(1, 31)}  # Conjunto de números naturales del 1 al 30
J = {x for x in I if x > 1 and all(x % i != 0 for i in range(2, int(x**0.5) + 1))}  # Buscamos las x en el conjunto I que cumplen con las condiciones de ser mayores que uno y que, al recorrer todos los numeros desde 2 a la raiz cuadrada del número, no haya ni uno solo que sea multiplo del numero
print("Conjunto I:", I)
print("Conjunto J (primos < 30):", J)
```

Ejercicio 5. Cuadrados perfectos menores que 50
Definir el conjunto K con los números naturales del 1 al 50 y luego obtener el
subconjunto L con los cuadrados perfectos.

```python

K = {x for x in range(1, 51)}  # Conjunto de números naturales del 1 al 50
L = {x for x in K if (x**0.5)%1 == 0 }  #Encontramos los numeros que al realizar su raiz cuadrada son enteros debido a que al ser divididos por uno no tiene residuo
print("Conjunto K:", K)
print("Conjunto L (cuadrados perfectos < 50):", L)
```

Ejercicio 6. Números que sean la suma de dos números consecutivos
Definir el conjunto M de con los números naturales del 1 al 100 y luego definir el
conjunto N con los números que son la suma de 2 números consecutivos.

```python
M = {x for x in range(1, 101)}  # Conjunto de números naturales del 1 al 100
N = {n + (n + 1) for n in M if n!=100}  # Este problema no lo entendi bien pero aqui se hace la suma de todos los n + n+1 del 1 al 100 excluyendo al 100 para que el resultado no se salga
print("Conjunto M:", M)
print("Conjunto N (suma de dos consecutivos):", N)
```

Ejercicio 7. Números que son múltiplos de 2 y 3 simultáneamente
Define el conjunto Q con los números del 1 al 30 y luego obtener el subconjunto
R con los números que son múltiplos de 2 y 3 a la vez.

```python
Q = {x for x in range(1, 31)}  # Conjunto de números del 1 al 30
R = {x for x in Q if x % 2 == 0 and x % 3 == 0}  # Obtenemos los numeros que son multiplos de 2 y de 3
print("Conjunto Q:", Q)
print("Conjunto R (múltiplos de 2 y 3):", R)
```

Ejercicio 8. Números que son factor de 100
Define el conjunto S con los números del 1 al 100 y luego obtener el subconjunto
T con los números que son factor de 100.

```python
# Ejercicio 8: Números que son factores de 100
S = {x for x in range(1, 101)}  # Conjunto de números del 1 al 100
T = {x for x in S if 100 % x == 0}  # Obtenemos todos los factores de 100
print("Conjunto S:", S)
print("Conjunto T (factores de 100):", T)
```
